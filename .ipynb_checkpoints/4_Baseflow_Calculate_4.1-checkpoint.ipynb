代码介绍  
import plotly.io as pio

这是引入 Plotly 的 I/O 接口模块（io = input/output）。

plotly.io 提供了一些全局的配置接口，比如：

默认渲染方式 (pio.renderers.default)

导出为静态图片 (pio.write_image)

保存为 HTML 文件 (pio.write_html)

Plotly 图表是 基于 JavaScript 的，Python 只是生成 JSON 配置。

所以在 Jupyter Notebook 里，要告诉 Plotly：用哪种方式把前端 JavaScript 嵌入并渲染出来。

这就叫 renderer（渲染器）。

常见渲染器有：

'notebook' → 在 Jupyter 里直接渲染（有时需要 nbextension 才能工作）

'notebook_connected' → 推荐的，自动加载必要的 JS 文件，兼容性更好

'vscode' → 在 VS Code 的 Jupyter Notebook 环境里使用

'browser' → 打开浏览器新标签页显示图表

'iframe' → 以 iframe 的方式嵌入

'png' → 直接显示静态 PNG 图片

from IPython.display import HTML

IPython.display 是 Jupyter 环境的一个工具包，用来在 Notebook 单元格里显示丰富内容（比如 HTML、图片、音频、视频等）。

HTML() 是其中的一个函数，可以把一段 HTML 代码嵌入 Notebook。

在 Plotly 里，我们常用 HTML(fig.to_html()) 的方式，把图表导出成 HTML 字符串，再直接在 Notebook 单元格中显示。

好处是：即使渲染器有问题，你也能强行在 Notebook 里显示完整交互图表。

代码介绍

fig.update_layout(...)

fig 是 Plotly 的图对象（go.Figure 或 px.scatter_map 返回的对象）。

update_layout 方法用于修改 图表布局（layout），包括标题、坐标轴、边距、颜色条、地图样式等。

它的参数是一个键值对配置，里面的每个键都对应 Plotly 的 layout 属性。

mapbox_style='open-street-map'

这行指定地图的底图（tiles）样式。

在 scatter_mapbox（或新版本 scatter_map）中，地图的背景是通过 Mapbox / MapLibre 的样式控制的。

'open-street-map' 表示使用 开源的 OpenStreetMap 作为底图，不需要任何 Token。

常见可选值有：

'open-street-map' → 标准 OSM 底图

'carto-positron' → 简洁的浅色底图（适合叠加彩色点/面）

'carto-darkmatter' → 深色底图（适合做夜景/对比强的地图）

'stamen-terrain' → 地形风格底图（山地、河流显著）

'white-bg' → 白色背景（只显示你的数据）

margin=dict(l=0, r=0, t=0, b=0)

margin 控制图表与画布边缘之间的空白。

dict(l=0, r=0, t=0, b=0) 表示：

l = left 左边距 0

r = right 右边距 0

t = top 上边距 0

b = bottom 下边距 0

margin = dict(l=80, r=80, t=20, b=50)调整画布与对话框的间距，适当的距离可以显示完全

```python
#单位换算
df_Site_No_extracted = df_Site_No[columns_to_extract]
df_Site_No_extracted['drain_area_va'] = df_Site_No_extracted['drain_area_va']*2.59

#site_no的地理信息处理
#重命名索引、列名，调整列的顺序
new_columns = {'dec_long_va': 'lon', 'dec_lat_va': 'lat', 'drain_area_va': 'area'}
df_Site_No_extracted_New_Index = df_Site_No_extracted.set_index('site_no').rename(columns = new_columns)[["lon", "lat", "area"]]

#取消索引列的列名
df_Site_No_extracted_New_Index.index.name = None
```

```python
BMNF = []#BMN：Best Method Name Mean Baseflow BFI
MBF  = []#Mean year baseflow
# year = list(range(2000,2021))
# MBF  = pd.DataFrame(columns = year)#Mean year baseflow
None_Site_No = []
n = 0
for csv in csv_files:
    site_no = csv.split('.')[0].split('_')[1]
    csv_path = os.path.join(Streamflow_Data_Path, csv)
    df_csv = pd.read_csv(csv_path)
    df_csv = df_csv[['dateTime', 'X_00060_00003']]
    new_columns_1 = {'dateTime': 'time', 'X_00060_00003': site_no}
    df_csv = df_csv.rename(columns = new_columns_1)
    #流量数据
    df_csv = df_csv.set_index('time')
    #地理数据
    df_loc = df_Site_No_extracted_New_Index.loc[[site_no]]
    #进行基流分割
    df_baseflow, df_bfi, df_est = baseflow.separation(df_csv, df_loc, return_bfi=True, return_kge=True)
    ##处理BMNF
    #方法名字处理
    best_method = df_est.idxmax(axis=1).reset_index()
    best_method.columns = ["site_no", "best_method"]
    best_method = best_method.set_index("site_no")
    best_method_name = best_method['best_method'].tolist()[0]
    #在取值前跳过结果站点
    if pd.notna(best_method_name):
        Best_KGE = round(df_est[best_method_name].tolist()[0],2)
        Best_BFI = round(df_bfi[best_method_name].tolist()[0],2)
        Baseflow_Mean = df_baseflow[best_method_name][site_no].mean().round(2)
        df_BF = df_baseflow[best_method_name]
        df_BF.index = pd.to_datetime(df_BF.index)
        df_BF['year'] = df_BF.index.year
        df_BF_grouped = df_BF.groupby(['year'])[site_no].mean().reset_index()
        df_BF_grouped = df_BF_grouped.set_index('year').T
        #数据信息处理
        BMNF.append({"site_no": site_no, "best_method": best_method_name, "mean_baseflow": Baseflow_Mean, "Best_BFI": Best_BFI, "Best_KGE":Best_KGE})   #逐一添信息到列表中
        ##处理MBF
        MBF.append(df_BF_grouped)
    else:
        None_Site_No.append(site_no)
        # Best_BFI = None
        # Baseflow_Mean = None
        # df_BF = df_baseflow[best_method_name]
        # df_BF.index = pd.to_datetime(df_BF.index)
        # df_BF['year'] = df_BF.index.year
        # df_BF_grouped = df_BF.groupby(['year'])[site_no].mean().reset_index()
        # df_BF_grouped = df_BF_grouped.set_index('year').T



df_BMNF = pd.DataFrame(BMNF)

df_MBF = pd.concat(MBF, axis=0)
```
